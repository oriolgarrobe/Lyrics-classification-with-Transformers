```python
import pandas as pd
import numpy as np
import zipfile
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt
import json

with open('data/lr_performance.json') as lr_dict:
    lr_performance = json.load(lr_dict)

with open('data/svm_performance.json') as svm_dict:
    svm_performance = json.load(svm_dict)
    
with open('data/bert_performance.json') as bert_dict:
    bert_performance = json.load(bert_dict)
```

```python
from helper_functions import plot_comp

# Create arrays for plotting Results

sample_size = ['1000', '2000', '4000', '6000', '10000', '15000', '20000', '40000']
class_size = ['2', '3', '4', '5', '6']      

# Create plot for each classification task based on number of classes

for nclass in class_size:
    
    # Allocate memory for f1 and time arrays
    f1 = np.zeros(24).reshape(3,8)
    time = np.zeros(24).reshape(3,8)

    counter = 0
    
    for size in sample_size:
        for model in range(3):
            if model==0:
                f1[model][counter]=lr_performance[size][nclass]['f1']
                time[model][counter]=lr_performance[size][nclass]['time']
            elif model==1:
                f1[model][counter]=svm_performance[size][nclass]['f1']
                time[model][counter]=svm_performance[size][nclass]['time']
            else:
                f1[model][counter]=bert_performance[size][nclass]['f1']
                time[model][counter]=bert_performance[size][nclass]['time']
        counter += 1
    
    # Function that shows and stores the plot for each number of classes
    plot_comp(f1, time, int(nclass))
```
